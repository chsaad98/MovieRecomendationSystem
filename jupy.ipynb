```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
from textblob import TextBlob

# Load the first CSV file
link_csv = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/link.csv')

# Impute missing values for the first CSV
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(link_csv.iloc[:, 1:3])
link_csv.iloc[:, 1:3] = imputer.transform(link_csv.iloc[:, 1:3])

print(link_csv.isnull().sum())

# Split the first CSV data into training and testing sets
x_train, x_test = train_test_split(link_csv, test_size=0.25, random_state=0)

print(x_train)
print(x_test)

# Plot the histogram for the first CSV
plt.hist(link_csv.iloc[:, 1:3], bins=10)
plt.title('Histogram of Link CSV Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Load the second CSV file
genome_scores = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/genome_scores.csv')

# Impute missing values for the second CSV
imputer.fit(genome_scores.iloc[:, 1:3])
genome_scores.iloc[:, 1:3] = imputer.transform(genome_scores.iloc[:, 1:3])

print(genome_scores.isnull().sum())

# Split the second CSV data into training and testing sets
y_train, y_test = train_test_split(genome_scores, test_size=0.25, random_state=0)

print(y_train)
print(y_test)

# Plot the histogram for the second CSV
plt.hist(genome_scores.iloc[:, 1:3], bins=10)
plt.title('Histogram of Genome Scores Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Load the third CSV file
tags = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/genome_tags.csv')

# Check for missing values
print(tags.isnull().sum())

# Impute missing values for 'tagId'
imputer_id = SimpleImputer(missing_values=np.nan, strategy='mean')
tags['tagId'] = imputer_id.fit_transform(tags[['tagId']])

# Fill missing values for 'tag' with 'Unknown'
tags['tag'].fillna('Unknown', inplace=True)

# Split the third CSV data into training and testing sets
tags_train, tags_test = train_test_split(tags, test_size=0.25, random_state=0)

# Display training and testing sets
print(tags_train)
print(tags_test)

# Plotting histograms for 'tagId'
plt.hist(tags['tagId'], bins=10)
plt.title('Histogram of Genome Tags Data')
plt.xlabel('Tag ID')
plt.ylabel('Frequency')
plt.show()

# Load the fourth CSV file
movies = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/movie.csv')

# Check for missing values
print(movies.isnull().sum())

# Impute missing values for 'title' and 'genres'
movies['title'].fillna('Unknown', inplace=True)
movies['genres'].fillna('Unknown', inplace=True)

# Create new features (Year and Number of Genres)
movies['year'] = movies['title'].str.extract(r'\((\d{4})\)')
movies['num_genres'] = movies['genres'].apply(lambda x: len(x.split('|')))

# Split the movie data into training and testing sets
movies_train, movies_test = train_test_split(movies, test_size=0.25, random_state=0)

# Display training and testing sets
print(movies_train)
print(movies_test)

# Plot the distribution of genres
genre_counts = movies['genres'].str.split('|', expand=True).stack().value_counts()

# Plot the genre distribution
plt.bar(genre_counts.index, genre_counts.values)
plt.title('Genre Distribution in Movies Data')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()

# Load the fifth CSV file
ratings = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/rating.csv')

# Check for missing values
print(ratings.isnull().sum())

# Ensure 'rating' is numeric
ratings['rating'] = pd.to_numeric(ratings['rating'], errors='coerce')

# Impute missing values for 'rating'
imputer_rating = SimpleImputer(missing_values=np.nan, strategy='mean')
ratings['rating'] = imputer_rating.fit_transform(ratings[['rating']])

# Split the rating data into training and testing sets
ratings_train, ratings_test = train_test_split(ratings, test_size=0.25, random_state=0)

# Display training and testing sets
print(ratings_train)
print(ratings_test)

# Plot the distribution of ratings
plt.hist(ratings['rating'], bins=10, edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Load the tag CSV file
tags = pd.read_csv('/Users/chaudhrysaad/Downloads/archive/tag.csv')

# Check for missing values
print(tags.isnull().sum())

# Check data types
print(tags.dtypes)

# Convert 'timestamp' to datetime
tags['timestamp'] = pd.to_datetime(tags['timestamp'])

# Display first few rows
print(tags.head())

# Sentiment analysis on tags
tags['sentiment'] = tags['tag'].apply(lambda x: TextBlob(x).sentiment.polarity if isinstance(x, str) else None)
print(tags[['tag', 'sentiment']].head())

# Plot the most common tags
tag_counts = tags['tag'].value_counts()
plt.figure(figsize=(10, 6))
tag_counts[:10].plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Tags')
plt.xlabel('Tags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Load and merge datasets
merged_data = pd.merge(ratings, movies, on='movieId')
merged_data = pd.merge(merged_data, tags, on=['movieId', 'userId'], how='left')

# Check the first few rows of the merged dataset
print(merged_data.head())

# Limit to top users and movies to reduce size
top_users = ratings['userId'].value_counts().head(1000).index  # Select top 1000 users
top_movies = ratings['movieId'].value_counts().head(1000).index  # Select top 1000 movies

# Filter merged_data
filtered_merged_data = merged_data[
    merged_data['userId'].isin(top_users) & merged_data['movieId'].isin(top_movies)
]

# Create the pivot table
movie_ratings = filtered_merged_data.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)

# Correlation analysis on numeric columns only
corr = filtered_merged_data.select_dtypes(include=[np.number]).corr()  # Only include numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Plot the distribution of ratings in merged data
plt.figure(figsize=(10, 6))
plt.hist(filtered_merged_data['rating'], bins=10, alpha=0.7, color='blue')
plt.title('Distribution of Ratings in Merged Data')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Collaborative filtering recommendation system
# Fit Nearest Neighbors model
model = NearestNeighbors(metric='cosine', algorithm='brute')
model.fit(movie_ratings)

# Recommend based on a user
user_index = 1  # Change to desired user index
distances, indices = model.kneighbors(movie_ratings.iloc[user_index, :].values.reshape(1, -1), n_neighbors=6)

print('Recommended movies for user:', user_index)
for i in range(1, len(indices.flatten())):
    movie_id = movie_ratings.columns[indices.flatten()[i]]
    movie_title = movies[movies['movieId'] == movie_id]['title'].values[0]
    print(f"Movie: {movie_title}, Distance: {distances.flatten()[i]}")

# Evaluate the recommendation system with RMSE
y_true = ratings_test['rating']
y_pred = ratings_train['rating'][:len(y_true)]  # Dummy prediction, replace with actual model
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"RMSE: {rmse}")
```
